```python
n=imu.shape[0]
m=enc.shape[0]

a_x=imu[:,2]
a_y=imu[:,3]
a_z=imu[:,4]

bias_x=np.mean(a_x[:9500])
bias_y=np.mean(a_y[:9500])

a_x=a_x-bias_x
a_y=a_y-bias_y

g=np.mean(a_z)
```

```python
def lp_filter(arr, alpha = 0.02):
    fil_arr = [arr[0]]
    for a in arr[1:]:
        fil_arr.append(alpha*a+(1-alpha)*fil_arr[-1])
    return fil_arr
```

```python
def hp_filter(arr, alpha = 0.02):
    fil_arr = [arr[0]]
    for i in range(arr.shape[0]):
        if i==0:
            continue
        fil_arr.append(alpha*fil_arr[-1]+(1-alpha)*(arr[i]-arr[i-1]))
    return fil_arr
```

```python
ax=lp_filter(a_x)
ay=lp_filter(a_y)
az=lp_filter(a_z)
wz=hp_filter(imu[:,7])
```

```python
plt.figure(figsize=(30,10))
plt.plot(range(n),imu[:,7])
plt.plot(range(n),wz)
```

```python
vel=[]

vel.append(np.array([0,0,0]))
state=np.array([0,0,0])

thet=0

P=np.eye(3)
P=P*0.07

sigma_u=np.eye(3)
sigma_u=sigma_u*0.027

R=np.eye(3)
R=R*0.01

for i in range(m):
    u=np.array([ax[i],ay[i],wz[i]])
    
    z=np.array([enc[i][2]*np.cos(thet),enc[i][2]*np.sin(thet),enc[i][3]])
    
    G=np.eye(3)
    G=G*0.01
    G[2][2]=1
    
    temp_x=np.array([state[0]+(0.01*u[0]),
                    state[1]+(0.01*u[1]),
                    u[2]])
    
    F=np.eye(3)
    F[2][2]=0
    
    Q=np.dot(G,sigma_u)
    Q=np.dot(Q,np.transpose(G))
    
    temp_p=np.dot(F,P)
    temp_p=np.dot(temp_p,np.transpose(F))
    temp_p=temp_p+Q
    
    H=np.eye(3)
    temp_z=np.dot(H,np.transpose(temp_x))
    
    K=np.dot(temp_p,np.transpose(H))
    temp=np.dot(H,temp_p)
    temp=np.dot(temp,np.transpose(H))
    temp=np.linalg.pinv(temp+R)
    K=np.dot(K,temp)
    
    P=np.dot((np.eye(3)-np.dot(K,H)),temp_p)
    state=temp_x+np.dot(K,(z-temp_z))
    thet=thet+(0.01*state[2])
    
    vel.append(state)
    
vel=np.array(vel)
```

```python
temp_s=[]
ph=0
temp_s.append(np.array([0,0]))

for i in range(m):
    temp_s.append(np.array([temp_s[-1][0]+(0.01*enc[i][2]*np.cos(ph)),temp_s[-1][1]+(0.01*enc[i][2]*np.sin(ph))]))
    ph=ph+(0.01*enc[i][3])
    
temp_s=np.array(temp_s)
```

```python
pose=[]
pose.append(np.array([0,0]))
for i in range(m):
    pose.append(np.array([pose[-1][0]+(0.01*vel[i,0]),pose[-1][1]+(0.01*vel[i,1])]))

pose=np.array(pose)
plt.figure(figsize=(10,10))
plt.scatter(pose[:,0],pose[:,1],s=0.1)
plt.scatter(temp_s[:,0],temp_s[:,1],s=0.1)
```
